** CELL 3 — Read data (only caption)**

```python
# === CELL 3: Read inputs with simplified ===
import csv, json, os
from typing import List, Dict

def read_csv(path: str) -> List[Dict]:
    rows = []
    with open(path, "r", newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for r in reader:
            cap  = (r.get("caption") or "").strip()
            simp = (r.get("simplified") or r.get("simplified_caption") or "").strip()
            if not cap:
                continue
            rows.append({
                "image_id": str(r.get("image_id")).strip(),
                "caption": cap,
                "simplified": simp,   # may be empty if column missing
            })
    return rows

def read_jsonl(path: str) -> List[Dict]:
    rows = []
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            obj = json.loads(line)
            cap  = (obj.get("caption") or "").strip()
            simp = (obj.get("simplified") or obj.get("simplified_caption") or "").strip()
            if not cap:
                continue
            rows.append({
                "image_id": str(obj.get("image_id")).strip(),
                "caption": cap,
                "simplified": simp,   # may be empty if key missing
            })
    return rows

def load_input(csv_path: str = None, jsonl_path: str = None) -> List[Dict]:
    if csv_path and os.path.exists(csv_path):
        return read_csv(csv_path)
    if jsonl_path and os.path.exists(jsonl_path):
        return read_jsonl(jsonl_path)
    raise FileNotFoundError("No input file found. Put CSV/JSONL in NLP_Project/inputs and set the path.")

rows = load_input(CSV_PATH, JSONL_PATH)
print(f"Loaded {len(rows)} rows")
rows[:2]
```

**CELL 4 — Prompt + generation (caption only)**

```python
# === CELL 4: Prompt for BOTH captions ===
SYSTEM_PROMPT = (
"You are a careful question writer.\n"
"You will get two short captions about the same image: a detailed caption and a simplified caption.\n"
"Write 1-3 diverse, unambiguous question–answer pairs that are strictly inferable from these captions.\n"
"\n"
"Rules:\n"
"- Use only facts supported by BOTH captions (or by the simplified caption when the detailed caption does not contradict it).\n"
"- Prefer specific details: counts, objects, attributes, relations, locations, actions.\n"
"- Avoid speculation and world knowledge beyond the text.\n"
"- Keep answers short (1–3 words) or 'yes/no'. Prefer non‑yes/no when possible (at most one yes/no).\n"
"\n"
"Example:\n"
'Detailed: \"A building with a black and gold clock on it."\n'
'Simplified: \"A building with a clock."\n'
"Pairs:\n"
'  - Question: What object is on the building?  → Answer: clock\n'
'  - Question: Where is the clock? → Answer: on the building\n'
'  - Question: What does the building have? → Answer: a clock\n'
'  - Question: Is there a clock on the building? → Answer: yes\n'
"\n"
"Output ONLY valid JSON in this schema:\n"
'{\n  \"pairs\": [\n    {\"question\": \"...\", \"answer\": \"...\"},\n    {\"question\": \"...\", \"answer\": \"...\"}\n  ]\n}\n'
"No extra text."
)

def build_user_prompt(caption: str, simplified: str) -> str:
    s = simplified or ""
    return f'Detailed: "{caption}"\nSimplified: "{s}"\n'

import torch, json

@torch.inference_mode()
def generate_pairs_from_both(
    caption: str,
    simplified: str,
    temperature=0.2, top_p=0.9, max_new_tokens=256, retries=2
):
    """Return up to 3 {'question': str, 'answer': str} using BOTH captions."""
    user_prompt = build_user_prompt(caption, simplified)
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": user_prompt},
    ]
    input_ids = tokenizer.apply_chat_template(
        messages, add_generation_prompt=True, return_tensors="pt"
    ).to(model.device)

    for _ in range(retries + 1):
        out = model.generate(
            input_ids=input_ids,
            max_new_tokens=max_new_tokens,
            do_sample=(temperature > 0.0),
            temperature=temperature,
            top_p=top_p,
            pad_token_id=tokenizer.eos_token_id,
        )
        text = tokenizer.decode(out[0][input_ids.shape[-1]:], skip_special_tokens=True).strip()
        try:
            s = text.find("{"); e = text.rfind("}")
            payload = text[s:e+1] if (s != -1 and e != -1 and e > s) else text
            data = json.loads(payload)
            pairs = data.get("pairs", [])
            cleaned = []
            yesno_used = 0
            for p in pairs:
                q = (p.get("question") or "").strip()
                a = (p.get("answer") or "").strip()
                if not q or not a:
                    continue
                # Enforce at most one yes/no
                if a.lower() in {"yes", "no"}:
                    if yesno_used >= 1:
                        continue
                    yesno_used += 1
                cleaned.append({"question": q, "answer": a})
            if 1 <= len(cleaned) <= 4:
                return cleaned[:3]
        except Exception:
            temperature = min(0.6, temperature + 0.2)
            top_p = min(0.95, top_p + 0.03)
            continue
    return []
```

**CELL 5 — Run for all rows & save outputs**

```python
# === CELL 5: Batch run over BOTH captions & save ===
OUT_JSONL = f"{OUTPUTS_DIR}/vqa_pairs_from_both.jsonl"
OUT_CSV   = f"{OUTPUTS_DIR}/vqa_pairs_from_both.csv"
OUT_JSON  = f"{OUTPUTS_DIR}/vqa_pairs_from_both.json"

all_records, bad = [], 0

for i, r in enumerate(rows, 1):
    img_id = r["image_id"]
    cap    = r["caption"]
    simp   = r.get("simplified", "")

    pairs  = generate_pairs_from_both(cap, simp, temperature=0.2)
    if not pairs:
        bad += 1
        continue

    for k, p in enumerate(pairs, 1):
        all_records.append({
            "question_id": f"{img_id}_{k}",
            "image_id": img_id,
            "question": p["question"],
            "answer": p["answer"],
        })

    if i % 25 == 0:
        print(f"[{i}/{len(rows)}] ok; {bad} with no valid JSON")

# Save JSONL
with open(OUT_JSONL, "w", encoding="utf-8") as f:
    for rec in all_records:
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")

# Save pretty JSON
with open(OUT_JSON, "w", encoding="utf-8") as f:
    json.dump(all_records, f, indent=2, ensure_ascii=False)
print("✅ Also saved pretty JSON to:", OUT_JSON)

# Save CSV
import csv
fieldnames = ["question_id", "image_id", "question", "answer"]
with open(OUT_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(all_records)

print(f"Done. Wrote {len(all_records)} QA rows. Skipped (no valid JSON): {bad}")
print("JSONL:", OUT_JSONL)
print("JSON: ", OUT_JSON)
print("CSV:  ", OUT_CSV)
```
